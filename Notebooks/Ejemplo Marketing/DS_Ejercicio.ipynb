## Procesamiento de Datos

```python
#Vamos a tratar los datos perdidos, que según el txt son las variables que dicen 'unknown'
df = df.replace( 'unknown', np.nan)
# Count the NaNs values to verify
print(df.isnull().sum()) 
```

```python
#Vamos a rellenar las variables perdidas con la variable mas frecuente de cada columna

for col in df.columns:
    # Check if the column is of object type
    if df[col].dtypes == 'object':
        # Impute with the most frequent value
        df[col] = df[col].fillna(df[col].value_counts().index[0])

# Count the NaNs values to verify
print(df.isnull().sum())  
```

```python
#Eliminamos filas duplicadas
df = df.drop_duplicates()
df.shape
```

```python
#Creamos un nuevo dataset sin la variable duration
df1 = df.drop('duration', 1)
df1.columns
```

```python
#Encoding de las variables categóricas.
#En este punto existen varias alternativas, get_dummies, one_hot_encoding, etc.
#Usaremos LabelEncoder.
from sklearn.preprocessing import LabelEncoder 

LE = LabelEncoder()

for i in df1.columns:
    if df1[i].dtypes == 'object':
    # Use LabelEncoder to do the numeric transformation
        df1[i]=LE.fit_transform(df1[i])
```

```python
#Creamos nuestro X e Y

X = df1.drop('y', 1)
y = df1['y']
```

## Machine Learning Model

Probaremos dos métodos muy famosos para tratar con problemas tabulares: Xgboost, y Lightgbm.
Dado que para modelos de árboles la data no normalizada no suele afectar, obviaremos este paso.
La métrica elegida para evaluar nuestro modelo será AUC_ROC, ya que la data esta muy desbalanceda y el Accuracy mucho no nos va a decir.

### Xgboost

```python
#Importamos las librerias necesarias y dividimos los datos en test y train
from sklearn import datasets
import xgboost as xgb
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.25)
```

```python
#Hacemos una primera prueba
from sklearn.metrics import roc_curve, auc, precision_recall_curve, classification_report, average_precision_score
from sklearn.calibration import CalibratedClassifierCV

clf = xgb.XGBClassifier(max_depth = 3,
                        learning_rate = 0.3,
                        objective = 'binary:logistic'
                        )  
clf.fit(X_train,
        Y_train
        )

Y_pred = clf.predict_proba(X_test)
y_true = np.array(Y_test)
y_scores = Y_pred[:, 1]

fpr, tpr, _ = roc_curve(Y_test, y_scores)
roc_auc = auc(fpr, tpr)
average_precision = average_precision_score(Y_test, y_scores)
precision, recall, _ = precision_recall_curve(Y_test, y_scores)

def plot_roc(fpr, tpr, roc_auc):
    plt.plot(fpr, tpr)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC curve (area = %0.6f)' % roc_auc)
    plt.legend(loc="lower right")
    plt.show()

plot_roc(fpr, tpr, roc_auc)
```

```python
#Probamos un Tuneo de Hyperparametros a ver si mejora.
#En este caso usamos RandomizedSearcCV para que tarde un poco menos en correr
from sklearn.model_selection import RandomizedSearchCV

clf = xgb.XGBClassifier(tree_method='gpu_hist', max_bin=16)
parameters = {
     "learning_rate"    : [0.05, 0.10, 0.15, 0.20, 0.25, 0.30 ] ,
     "max_depth"        : [ 3, 4, 5, 6, 8, 10, 12, 15],
     "gamma"            : [ 0.0, 0.1, 0.2 , 0.3, 0.4 ],
     }

grid = RandomizedSearchCV(clf,
                    parameters, n_jobs=4,
                    scoring="roc_auc",
                    cv=3)

best_model = grid.fit(X_train, Y_train)
best_model.best_params_
```

Pasos a seguir:
